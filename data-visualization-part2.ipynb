```python

import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('raw_data1.csv') 
# print(df.columns)
# age_df = df.iloc[:]['Age Bracket'].dropna()
age_df = df['Age Bracket'].dropna()
age_list = list(age_df)
for i, v in enumerate(age_list):
  if len(v.split('-'))==2:
    v = (float(v.split('-')[0]) + float(v.split('-')[1]))/2 # hard coded mean calculation
  age_list[i] = float(v)

x_data = []
y_data = []
for i in range(0, 100):
  x_data.append(i)
  y_data.append(age_list.count(i))

plt.figure(figsize=(60, 30))
plt.bar(x_data, y_data)
plt.plot(x_data, y_data, color='red', linewidth=4)
# plt.legend(['line graph', 'bar graph'], prop={'size':40}) #labels with style
plt.xticks(size=40, color='black')
plt.yticks(size=40, color='black')
plt.xlabel('Age', size=40, color='green')
plt.ylabel('Number of confirmed cases', size=40, color='green')
plt.title('covid19 confirmed cases vs age ', size=40, color='green')
plt.show()
```
