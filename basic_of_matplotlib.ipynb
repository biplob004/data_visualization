```python


# basic of matplot lib

import matplotlib.pyplot as plt

plt.figure(figsize=(16,9))
plt.xlabel("x-label", size=20)
plt.ylabel("y-label", size=20)
plt.xticks(size=20)
plt.yticks(size=20)
plt.title("title", size=30)
# plt.plot(['y1','y2','y3','y4','y5'],[7,8,9,10,30]) # line graph
plt.bar(['y1','y2','y3','y4','y5'],[7,8,9,10,30], color='purple') # bar graph
plt.bar(['y1','y2','y3','y4','y5'],[2,4,7,13,15], color='red') # bar graph
plt.plot(['y1','y2','y3','y4','y5'],[7,8,9,10,30], linestyle='--', color='blue') # color and style
plt.axhline([15], color='black', linestyle='--') # middle line 
plt.legend(['legend1', 'legend2', 'legend3'], prop={'size':20}) #labels with style
plt.show()

```
